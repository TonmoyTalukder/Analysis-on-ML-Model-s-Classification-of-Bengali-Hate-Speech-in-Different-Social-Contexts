```python
from bltk.langtools.banglachars import (punctuations)

# print(f'Punctuation marks: {punctuations}')
bn_punctuations = punctuations
```

```python
def basic_clean(words):
  wnl = nltk.stem.WordNetLemmatizer()

  return [wnl.lemmatize(word) for word in words if word not in bn_punctuations ] 
```

```python
a = 'আমি ভাত খাই। '
basic_clean(a.split())
```

```python
import string
def stopwords_clean(text):
  reText = []
  text = text.translate(str.maketrans('', '', string.punctuation))
  newtext = remove_stopwords(text, bnstopwords)
  for i in newtext:
    if i != "।":
      reText.append(i)
  return reText
```

```python
import string
sen_list =  list(df["sentence"])
df_words = []
for i in sen_list:
  df_words = df_words + stopwords_clean(i)
```
